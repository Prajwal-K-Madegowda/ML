# 1. Linear Regression

## 1. The Core Idea

Linear regression is about finding a **straight-line relationship** between two things:

- An **input** (something you know)
- An **output** (something you want to predict)

That straight line is the simplest possible rule connecting them.

---

## 2. Real-world example

Imagine you’re looking at people’s **height** and **weight**.

- Generally, taller people weigh more.
- If we put all the data points (height vs. weight) on a graph, they form a kind of upward trend.

Linear regression says:

> “Let’s draw the best straight line through those points, so we can predict someone’s weight just by knowing their height.”
> 

---

## 3. The Simple Math (in plain words)

The straight line is usually written as:

                            y=mx+c 

Where:

- **x** = input (e.g., height)
- **y** = output we want to predict (e.g., weight)
- **m** = slope of the line (how much y changes when x increases by 1)
- **c** = intercept (where the line starts when x = 0)

So the line might look like:

> Weight = 2 × Height + 30
> 

That means: for each extra cm of height, the person’s weight increases by about 2 kg.

---

## 4. How the computer finds the line

The computer doesn’t just draw any line.

It tries to minimize the **error** (the difference between the actual values and the predicted ones).

This is done using a method called **“least squares”** (it finds the line where the total squared error is the smallest).

---

## 5. Where it’s used

- Predicting **house prices** from size
- Estimating **sales** based on advertising spend
- Predicting **temperature** tomorrow from today’s temperature

Basically, whenever you want to **predict a number** and you think there’s a roughly straight-line trend.

---

👉 In short: **Linear regression = finding the best straight-line rule to predict one thing from another.**

## CODE

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Independent variable (Years of Experience)
# Must be a 2D array for scikit-learn
X = np.array([[1], [2], [3], [4], [5]])

# Dependent variable (Salary)
y = np.array([30000, 45000, 58000, 70000, 82000])

# Create a Linear Regression model
model = LinearRegression()

# Train the model using the provided data
model.fit(X, y)

# Calculate R² score to evaluate model performance
# R² close to 1 indicates a good fit
score = model.score(X, y)
print(f"Model Score: {score:.5f}\n")

# Calculate Root Mean Squared Error (RMSE)
# RMSE close to 0 indicates low prediction error
predictions = model.predict(X)
rmse = np.sqrt(mean_squared_error(y, predictions))
print(f"Root Mean Squared Error: {rmse:.2f}\n")


# Continuously take user input until user types 'stop'
while True:
    user_input = input("Enter years of experience to predict salary (or type 'stop' to exit): ")

    # Exit condition
    if user_input.lower() == 'stop':
        break

    try:
        # Convert input to float
        years_experience = float(user_input)

        # Convert input into 2D array for prediction
        predicted_salary = model.predict(np.array([[years_experience]]))

        # Print predicted salary formatted to 2 decimal places
        print(
            f"Predicted Salary for {years_experience} years of experience: "
            f"${predicted_salary[0]:.2f}\n"
        )

    except ValueError:
        # Handles non-numeric input
        print("Invalid input. Please enter a numeric value for years of experience.\n")
```
