### Introduction to Scikit-Learn



Scikit-Learn is a toolkit for traditional machine learning (e.g., regression, classification, clustering, PCA) without deep learning. 

It focuses on workflows like data loading, splitting, preprocessing, model training, evaluation, hyperparameter tuning, and pipelines.

### Basic Example: Breast Cancer Classification

#### 1. Importing Libraries

```python
from sklearn.datasets import load_breast_cancer 
# this is a dataset of breast cancer 

from sklearn.model_selection import train_test_split
# this is a function that splits the data into training and testing sets

from sklearn.preprocessing import StandardScaler 
# this is a function that standardizes the data

from sklearn.neighbors import KNeighborsClassifier 
# this is a function that implements the K-Nearest Neighbors algorithm
```

- **`load_breast_cancer`**: loads a built-in dataset of breast cancer measurements (features like tumor size, texture, etc. + target whether malignant or benign).
- **`train_test_split`**: splits the dataset into training and test sets.
- **`StandardScaler`**: standardizes features (mean=0, variance=1) so that all variables are on the same scale.
- **`KNeighborsClassifier`**: the ML algorithm used for classification.

#### 2. Load Dataset

##### way 1

```python
data=load_breast_cancer()
data
```

```python
X=data.data
y=data.target
print(X)
print(y)
```

#### way 2

```python
data=load_breast_cancer(as_frame=True).frame 
# this is a function that loads the breast cancer dataset and returns a pandas dataframe
```

```python
data.head()
```

```python
# Load data
X, y = load_breast_cancer(return_X_y=True)

# this is a function that loads the breast cancer dataset and returns a numpy array of features and a numpy array of labels
```

- `X`: input features (numeric tumor measurements, shape ≈ 569 samples × 30 features).
- `y`: target labels (0 = malignant, 1 = benign).
- `return_X_y=True`: directly returns `(X, y)` instead of a dictionary-like object.

#### 3. Split into Train/Test

```python
# Split data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

# this is a function that splits the data into training and testing sets
#test_size=0.2 means that 20% of the data will be used for testing and 80% will be used for training
```

- Splits the dataset into:
    - **80% training data** → `X_train`, `y_train`
    - **20% testing data** → `X_test`, `y_test`
- This ensures the model is trained on one part and evaluated on unseen data.

#### 4. Feature Scaling

```python
# Scale data

scalar= StandardScaler()    
#StandardScaler() is a function that standardizes the data

X_train_scaled = scalar.fit_transform(X_train) 
# calculates the mean/variance of training data and scales it.

X_test_scaled = scalar.transform(X_test)
 #scales test data using the same training mean/variance (to avoid data leakage).
```

- KNN is **distance-based**, so scaling is crucial.
- `fit_transform(X_train)`: calculates the mean/variance of training data and scales it.
- `transform(X_test)`: scales test data using the **same training mean/variance** (to avoid data leakage).

#### 5. Train the Model

```python
# Train model

knn = KNeighborsClassifier(n_neighbors=15)
 # this is a function that implements the K-Nearest Neighbors algorithm

knn.fit(X_train_scaled, y_train) 
# this is a function that fits the K-Nearest Neighbors algorithm to the training data
```

- Creates a KNN classifier (by default, `n_neighbors=5`).
- Fits it on the **scaled training data**.

How KNN works:

- For a new test sample, it looks at the closest `k` neighbors in the training set.
- Assigns the majority class label among them.

#### 6. Evaluate the Model

```python
# Evaluate
accuracy = knn.score(X_test_scaled, y_test)
# this is a function that evaluates the K-Nearest Neighbors algorithm on the test data
print(f"Accuracy: {accuracy}") 
```

- `.score()` gives accuracy = **(correct predictions / total predictions)**.
- Prints the classification accuracy on test data.

**Accuracy based on number of k's**

- #Accuracy: 0.9473684210526315 =15

##### Make prediction

```python
# Make predictions
prediction = knn.predict([X_test_scaled[18]])
print(f"Prediction: {prediction[0]}")
```

##### Synthetic Data Example:

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

- **`make_blobs`** → a function from scikit-learn that generates *synthetic (fake)* data for clustering or classification.
    - It creates groups (called *blobs*) of points around some centers.
- **`matplotlib.pyplot`** → a plotting library used to visualize data.

```python
X, y = make_blobs(n_samples=500, centers=5,random_state=42)

```

| Parameter | Meaning |
| --- | --- |
| `n_samples=500` | Create 500 data points. |
| `centers=5` | There will be **5 clusters (groups)** of points. |
| `random_state=42` | Ensures results are the same each time you run it (for reproducibility). |


**Outputs:**

- `X` → coordinates of the 500 points (a NumPy array with shape `(500, 2)` → 500 rows, 2 columns for X and Y coordinates).
- `y` → labels for each point (0–4) showing which cluster that point belongs to.

Example:
- X[0] → [1.23, 4.56]  # position of one point
- y[0] → 3             # belongs to cluster number 3


```python
plt.scatter(X[:, 0], X[:, 1], c=y) 
plt.show()
```

- `X[:, 0]` → all x-axis values
- `X[:, 1]` → all y-axis values
- `c=y` → color each point based on its cluster label (`y`), so each cluster appears in a different color
- `plt.show()` → displays the plot

You’ll see **five colored blobs** of points — each one is a cluster.

**Why we use it**
> This is a quick way to create and visualize **clustered data**, often used for:

- Testing clustering algorithms (like K-Means).
- Demonstrating how algorithms find patterns in unlabeled data.

#### Stratified Shuffle Split

**Why Stratified Splitting?**

Regular train-test split can create imbalanced distributions where training and testing sets have different class proportions. Stratified splitting ensures similar class ratios in both sets

```python
from sklearn.model_selection import StratifiedShuffleSplit
import numpy as np

# Create stratified split
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42) 
# this is a function that splits the data into training and testing sets while preserving the class distribution

for train_idx, test_idx in split.split(X, y):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

# Check class distribution
print("Training distribution:", np.bincount(y_train))
print("Testing distribution:", np.bincount(y_test))
```

**Mathematical Background:**

- **StandardScaler:** Xscaled=$X−μσX_{scaled} = \frac{X - \mu}{\sigma}Xscaled=σX−μ$ where $μ$ is mean and $σ$ sigma σ is standard deviation
- **MinMaxScaler:** Xscaled=$X−XminXmax−XminX_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}Xscaled=Xmax−XminX−Xmin$

**When to Scale:**

- **Required for:** Distance-based algorithms (KNN, SVM, Linear Regression)
- **Optional for:** Tree-based algorithms (Random Forest, Decision Trees)

```python
from sklearn.preprocessing import OrdinalEncoder

columns_to_encode = ['lug_boot', 'safety']

# For features with natural ordering (small < medium < big)
encoder = OrdinalEncoder(categories=[['small', 'med', 'big'], 
                                   ['low', 'med', 'high']])
data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])

# Reverse transformation
original_data = encoder.inverse_transform(data[columns_to_encode])
```

One-Hot Encoding (for nominal categories):

```python
data=fetch_openml('adult', as_frame=True).frame
```

```python
data.sample(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# For features without ordering (occupation, race, country)
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoded_values = encoder.fit_transform(data[['occupation', 'race']])
new_columns = encoder.get_feature_names_out(['occupation', 'race'])


# Create DataFrame with encoded features
df_encoded = pd.DataFrame(encoded_values, columns=new_columns, index=data.index)

# Combine with original data (dropping encoded columns)
data_final = pd.concat([data.drop(['occupation', 'race'], axis=1), df_encoded], axis=1)
```
